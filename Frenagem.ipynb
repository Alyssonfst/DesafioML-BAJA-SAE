```python
# Remover linhas onde ambas as colunas têm valores nulos
base_dados = base_dados.dropna(subset=['CD_Distancia de frenagem de 35 a 0 em asfalto', 'CD_Distancia de frenagem de 35 a 0 em terra batida'], how='all')

base_dados.info()

print(base_dados.info())
valores_nulos_por_coluna = base_dados.isnull().sum()

print(valores_nulos_por_coluna)
```

```python
# Calcular as médias das distâncias de frenagem em asfalto e terra
media_asfalto = base_dados['CD_Distancia de frenagem de 35 a 0 em asfalto'].mean()
media_terra = base_dados['CD_Distancia de frenagem de 35 a 0 em terra batida'].mean()

# Calcular a diferença relativa entre as distâncias de frenagem em asfalto e terra
diferenca_relativa = ((media_asfalto - media_terra) / media_terra) * 100  # Em percentual

print(f"A distância média de frenagem em asfalto é aproximadamente {diferenca_relativa:.2f}% melhor em relação à terra.")
```

```python
# Calcular a média dos valores não nulos em 'CD_Distancia de frenagem de 35 a 0 em asfalto'
media_asfalto = base_dados['CD_Distancia de frenagem de 35 a 0 em asfalto'].mean()

# Aplicar um aumento de 26,93% na média de asfalto para preencher os valores faltantes em terra batida
aumento = 1.2693  # Aumento de 26,93%
base_dados['CD_Distancia de frenagem de 35 a 0 em terra batida'].fillna(media_asfalto * aumento, inplace=True)

# Calcular a média dos valores não nulos em 'CD_Distancia de frenagem de 35 a 0 em terra batida'
media_terra = base_dados['CD_Distancia de frenagem de 35 a 0 em terra batida'].mean()

# Aplicar uma redução de 26,93% na média de terra para preencher os valores faltantes em asfalto
reducao = 0.7393  # Redução de 26,93%
base_dados['CD_Distancia de frenagem de 35 a 0 em asfalto'].fillna(media_terra * reducao, inplace=True)

base_dados
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train = np.array(base_dados.iloc[:, np.r_[1,3,4,5,7]])
y_train = np.array(base_dados['P_Frenagem'])

#x_train
#y_train

x_train, x_val,y_train, y_val = train_test_split(x_train,y_train,train_size=0.6,random_state=15)

modelo_rf = RandomForestClassifier(n_estimators=200, max_depth=12, random_state=12)  # Ajuste os parâmetros conforme necessário

modelo_rf.fit(x_train, y_train)

predicoes = modelo_rf.predict(x_val)

acuracia = accuracy_score(y_val, predicoes)
print(f'Acurácia do modelo de Random Forest: {acuracia:.2f}')
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Calcula a matriz de confusão
matriz_confusao = confusion_matrix(y_val, predicoes)

# Defina as classes reais e previstas
classes = ['1', '2', '3', '4', '5']

# Plotar a matriz de confusão usando seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_confusao, annot=True, cmap='Reds', fmt='g', cbar=False, xticklabels=classes, yticklabels=classes)

plt.title('Matriz de Confusão Frenagem')
plt.xlabel('Valores Previstos')
plt.ylabel('Valores Reais')
plt.show()
```

```python
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus

# Substitua o número do índice (0 neste caso) pelo número da árvore que deseja visualizar
estimator = modelo_rf.estimators_[0]

# Exporte a árvore para um arquivo .dot
export_graphviz(estimator, out_file='tree.dot',
                feature_names=['Diametro Disco traseiro', 'Diametro Disco frente', 'Dis.Frenagem asfalto','Dis.Frenagem terra', 'Massa Total'],  # Substitua pelos nomes das suas colunas
                class_names=['1', '2', '3', '4', '5'],  # Substitua pelas classes do seu modelo
                rounded=True, proportion=False,
                precision=2, filled=True)

# Converta o arquivo .dot para um formato legível
graph = pydotplus.graph_from_dot_file('tree.dot')

# Mostre a árvore
Image(graph.create_png())
```
